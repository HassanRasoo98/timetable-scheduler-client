```python
import pandas as pd

data = {
    'Day': ['Tuesday.xlsx', 'Monday.xlsx', 'Thursday.xlsx', 'Monday.xlsx', 'Wednesday.xlsx',
            'Monday.xlsx', 'Wednesday.xlsx', 'Thursday.xlsx', 'Tuesday.xlsx'],
    'Time': ['02:30-03:50', '01:00-02:20', '01:00-02:20', '01:00-02:20', '01:00-02:20',
             '03:55-05:15', '03:55-05:15', '11:30-12:50', '11:30-12:50'],
    'Class': ['A-108', 'A-301', 'A-301', 'A-108', 'B-230', 'C-408', 'B-230', 'A-108', 'A-108'],
    'Subject': ['Fund of NLP (AI-JK)', 'PPIT (AI-J)', 'PPIT (AI-J)', 'Art Neural Net (AI-J)',
                'Art Neural Net (AI-J)', 'Info Sec (AI-J)', 'Info Sec (AI-J)', 'Game Theory (AI-JK)',
                'Game Theory (AI-JK)'],
}

df = pd.DataFrame(data)
df[['Start_Time', 'End_Time']] = df['Time'].str.split('-', expand=True)
df = df[['Day', 'Time', 'Class', 'Subject', 'Start_Time', 'End_Time']]
# df['Time'] = pd.to_datetime(df['Time'], format='%I:%M%p')

df['Start_Time'].tolist()
```

```python
import re

def extract_numbers(input_string):
    # Use regex to find all numeric values in the string
    numbers = re.findall(r'\d+', input_string)
    return [int(number) for number in numbers]

# Example usage:
input_string = "The price of the product is $25 and the quantity is 10."
result = extract_numbers(input_string)
print(result)
```

```python
def format_time(time):
    hour = extract_numbers(time)[0]
    
    if hour >= 8 and hour < 12:
        return time + ' AM'
        
    else:
        return time + ' PM'
```

```python
df['Start_Time'] = df['Start_Time'].apply(format_time)
df['End_Time'] = df['End_Time'].apply(format_time)

df
```

```python
# Assuming 'Time' is the column containing time in 12-hour format without AM/PM
# Convert 'Time' to datetime objects
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='%I:%M %p')

# Sort the DataFrame by the 'Time' column
df_sorted = df.sort_values(by='Start_Time')
df_sorted
```
